```python
# looking for missing values
missing_in_wei = sales_df["Item_Weight"].isna()
#missing_in_ou_siz = dales_df["Outlet_Size"].isna()
sales_df.fillna(0)
```

```python
mean_avg = float("%.3f "%sales_df["Item_Weight"].mean())
mean_avg

item_weigh_lis = sales_df.loc[:, "Item_Weight"].fillna(mean_avg)
item_weigh_lis
sales_df.update({"Item_Weight" : item_weigh_lis})

```

```python
outlet_old_lis = sales_df["Outlet_Size"].fillna(0)
outlet_new_lis = []

for i in range(len(outlet_old_lis)):
    
    if outlet_old_lis[i] == 0 :
          if sales_df["Outlet_Location_Type"][i] == "Tier 3":
              outlet_new_lis.append("High")
          elif sales_df["Outlet_Location_Type"][i] == "Tier 2":
              outlet_new_lis.append("Small")
          else:
              outlet_new_lis.append("Medium")
    else :
      outlet_new_lis.append(outlet_old_lis[i])
sales_df.update({"Outlet_Size": outlet_new_lis})
sales_df
```
